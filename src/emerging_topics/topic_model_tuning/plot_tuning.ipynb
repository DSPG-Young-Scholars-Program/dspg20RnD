## Plot coherence and perplexity results (LDA only)

```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

### LDA plots

```python
c1 = pd.read_pickle("./results/lda_c.pkl")
c2 = pd.read_pickle("./results/lda_c2.pkl")
c3 = pd.read_pickle("./results/lda_c3.pkl")

p1 = pd.read_pickle("./results/lda_p.pkl")
p2 = pd.read_pickle("./results/lda_p2.pkl")
p3 = pd.read_pickle("./results/lda_p3.pkl")
```

```python
coherence = list(c1["iteration 0"]) + list(c2["iteration 0"]) + list(c3["iteration 0"])
perplexity = list(p1["iteration 0"]) + list(p2["iteration 0"]) + list(p3["iteration 0"])
```

```python
# plot results

n_topics = list(range(5,131,5)) + [140, 150, 175, 200]

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(n_topics, coherence, marker='o')
plt.title("LDA")
plt.xlabel("Number of Topics")
plt.ylabel("$C_{V}$ Topic Coherence")
#plt.legend(("coherence"), loc='best')

plt.subplot(1,2,2)
plt.plot(n_topics, perplexity, marker='o')
plt.title("Dataset 2008-2019")
plt.xlabel("Num Topics")
plt.ylabel("Perplexity")
#plt.legend(("perplexity"), loc='best')

plt.subplots_adjust(wspace=0.3)
#plt.tight_layout()
plt.show()
```
